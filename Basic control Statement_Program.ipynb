```python
#Largest of Three Numbers
a = 10
b = 25
c = 15

if a > b and a > c:
    print("A is largest")
elif b > c:
    print("B is largest")
else:
    print("C is largest")
```

```python
while True:
    print("1.Add 2.Subtract 3.Multiply 4.Exit")
    choice =int(input("Enter choice:"))

    if choice == 1:
        a = int(input("Enter a: "))
        b = int(input("Enter b: "))
        print("Sum =", a + b)

    elif choice == 2:
        a = int(input("Enter a: "))
        b = int(input("Enter b: "))
        print("Difference =", a - b)

    elif choice == 3:
        a = int(input("Enter a: "))
        b = int(input("Enter b: "))
        print("MUL=",a*b)

    elif choice == 4:
        break
```
