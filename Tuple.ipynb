```python
#Concatenation of Tuples
tup1 = (0, 1, 2, 3)
tup2 = ('Hope', 'AI', 'Tuple')

tup3 = tup1 + tup2
print(tup3)
```

```python
#Slicing of Tuple

tup = tuple('HOPEAI')

# Removing First element
print(tup[1:])

# Reversing the Tuple
print(tup[::-1])

# Printing elements of a Range
print(tup[4:9])
```

```python
# Tuple Unpacking with Asterisk (*)

tup = (1, 2, 3, 4, 5)
a, *b, c = tup

print(a) 
print(b) 
print(c)
```
