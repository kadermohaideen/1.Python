```python
bmi = int(input("Enter the BMI Index: "))

if bmi <= 19:
    print("Underweight")
elif 19 < bmi <= 25:
    print("Normal")
elif 25 <= bmi <= 30:
    print("Overweight")
else:
    print("Obese")
```
