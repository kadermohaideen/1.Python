```python
# Create a function that checks whether the given number is Odd or Even
class OddEven():
    def OddEven():
        nameData=int(input("Enter a number"))
        if (nameData % 2 != 0):
            print(nameData,"is Odd number");
        else:
            print(nameData,"is Even number");
```

```python
OddEven.OddEven()
```

```python
## Create a function that tells elegibility of marriage for male and female according to thei
##r age limit like 21 for male and 18 for female
class ElegiblityForMarriage:
    def Eligible():
        gender = input("Enter your Gender: ")
        age = int(input("Your age: "))

        if gender == "Male" and age >= 23:
            print("ELIGIBLE")
        elif gender == "Female" and age >= 18:
            print("ELIGIBLE")
        else:
            print("NOT ELIGIBLE")


ElegiblityForMarriage.Eligible()
```

```python
class FindPercent:

    def percentage(self):
        subject1 = 98
        subject2 = 87
        subject3 = 95
        subject4 = 95
        subject5 = 93

        nums = [subject1, subject2, subject3, subject4, subject5]

        total = sum(nums)
        avg = total / len(nums)

        print("Total =", total)
        print("Average =", avg)

FindPercent().percentage()
```

```python
class Triangle:

    def triangle(self):
        height = 32
        breadth = 34

        print("Area formula : (Height * Breadth) / 2")
        area = 0.5 * height * breadth
        print("Area of Triangle :", area)

        height1 = 2
        height2 = 4
        breadthval = 4

        print("Height1 :", height1)
        print("Height2 :", height2)
        print("Breadth :", breadthval)

        perimeter = height1 + height2 + breadthval
        print("Perimeter formula = Height1 + Height2 + Breadth")
        print("Perimeter of Triangle =", perimeter)
Triangle().triangle()
```
